 **What is OCR** (Optical Character Recognition)?
OCR is technology that reads text from images or photos and turns it into real text you can copy and edit. For example, when you scan a book page, OCR converts the picture into actual readable text instead of just an image.


**Q2: Why is Urdu OCR harder than English OCR?**
Urdu letters connect together in a cursive style and change shape depending on their position in a word, similar to joined handwriting. English letters, on the other hand, stay separate and keep the same shape. Urdu is also written right-to-left, and there is much less training data available for Urdu compared to English.


**Q3: What are 2 real-world situations where Urdu OCR would be useful?**
Urdu OCR can be used to scan old Urdu books and newspapers and convert them into digital, searchable text for archiving. It can also be used to automatically read Urdu signboards or official documents, which is helpful for translation apps or digitizing government paperwork.

```python
!pip install Pillow arabic-reshaper python-bidi

from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display
import os

# Sample Urdu sentences to generate
urdu_texts = [
    "اردو ایک خوبصورت زبان ہے",
    "پاکستان ہمارا وطن ہے",
    "یہ ایک مصنوعی ذہانت کا منصوبہ ہے",
    "میں کمپیوٹر سائنس کی طالبہ ہوں",
    "کتاب پڑھنا بہت اچھی عادت ہے",
    "سورج مشرق سے طلوع ہوتا ہے",
    "علم حاصل کرنا ہر مسلمان پر فرض ہے",
    "دوستی ایک انمول رشتہ ہے",
    "صبر کا پھل میٹھا ہوتا ہے",
    "محنت میں عظمت ہے",
    "وقت کی قدر کرنی چاہیے",
    "خوشی بانٹنے سے بڑھتی ہے",
    "سچائی ہمیشہ کامیاب ہوتی ہے",
    "درخت لگانا ایک نیک عمل ہے",
    "پانی زندگی کے لیے ضروری ہے",
    "اساتذہ کا احترام کرنا چاہیے",
    "صحت ہی اصل دولت ہے",
    "نظم و ضبط کامیابی کی کنجی ہے",
    "خواب دیکھنا اور پورا کرنا",
    "امید پر دنیا قائم ہے"
]

os.makedirs('/content/drive/MyDrive/data/raw', exist_ok=True)
for i, text in enumerate(urdu_texts):
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)
    img = Image.new('RGB', (400, 80), color='white')
    draw = ImageDraw.Draw(img)
    draw.text((10, 20), bidi_text, fill='black')
    img.save(f'/content/drive/MyDrive/data/raw/urdu_{i+1}.png')
    print(f'Generated: urdu_{i+1}.png -- {text}')

print('Done! Check data/synthetic/ folder')
```
